```python
import sys
import argparse
import json
import urllib
import requests
import pandas as pd
from pandas import json_normalize
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
def capital(x):
    '''
    Sacamos la altura segun poquemon
    '''
    url = f"https://restcountries.eu/rest/v2/name/{x}?fullText=true"
    response = requests.get(url)
    results = response.json()
    all_ids = [x["capital"] for x in results]
    return all_ids
```

```python
capital('spain')
```
